# Subsetting data


Subsetting release months in ascending order

```python
months_df = df.sort_values(by='released_month')
months_df.head(10)
months_df
```

# Visualizing data


Released month and music mode(major/minor)

```python
plt.bar(months_df['released_month'], months_df['mode'])
plt.xlabel('Released Month')
plt.ylabel('Music Mode')
plt.title('Released Month and Music Mode')

plt.show()
```
